```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
energies = [0.193,	0.133,	0.111,	0.103,	0.063,	0.062,	0.044,	0.046,	0.038]
eRatios = [4.09, 2.21, 1.48, 1.30, 1.44, 1.44, 1.33, 1.30, 1.00]
# plot the timeseries on the same subplot
plt.rcParams.update({'axes.labelsize': 18,
                    'xtick.labelsize': 16,
                    'xtick.major.width': 1,
                    'ytick.labelsize': 16,
                    'ytick.major.width': 1,
                    'legend.fontsize': 16,
                    'font.size': 16,
                    'font.family': 'sans-serif',
                    'font.sans-serif': 'Arial',
                    'mathtext.fontset': 'dejavusans',
                    'axes.linewidth': 1,
                    'lines.linewidth': 1.,
                    'lines.markersize': 1.})
fig, ax = plt.subplots(figsize=(6, 5))
ax.scatter(energies[0:2], eRatios[0:2], c = ['black', 'slategray'], alpha = 0.75, marker='D', s = 50)
ax.scatter(energies[2:4], eRatios[2:4], c = ['black', 'slategray'], alpha = 0.75, marker='s', s = 50)
ax.scatter(energies[4:6], eRatios[4:6], c = ['black', 'slategray'], alpha = 0.75, marker='X', s = 50)
ax.scatter(energies[6:8], eRatios[6:8], c = ['black', 'slategray'], alpha = 0.75, marker='P', s = 50)
ax.scatter(energies[8], eRatios[8], c = 'black', alpha = 0.75, marker='o', s = 50)
ax.set_xlabel("Average Energy Charge [$/kWh]")
ax.set_ylabel("Ratio of On : Off Peak Charge [-]")
ax.set_xlim(0.0, 0.20)
ax.set_ylim(0.0, 5.)
ax.set_xticks(np.arange(0.0, 0.25, 0.05))
ax.set_yticks(np.arange(0.0, 6., 1))
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: '{:.2f}'.format(x)))
fig.tight_layout()
fig.savefig('energyCharge.pdf', dpi=300, bbox_inches='tight')
fig.savefig('energyCharge.svg', dpi=300, bbox_inches='tight')
fig.savefig('energyCharge.png', dpi=300, bbox_inches='tight')
```

```python
demands = [10.5, 9.27, 47.50, 22.48, 12.47, 12.47, 44.25, 18.17, 2.33]
dRatios = [1.32, 1.15, 2.29, 1.09, 2.34, 2.34, 2.66, 4.32, 1.00]
# plot the timeseries on the same subplot
plt.rcParams.update({'axes.labelsize': 18,
                    'xtick.labelsize': 16,
                    'xtick.major.width': 1,
                    'ytick.labelsize': 16,
                    'ytick.major.width': 1,
                    'legend.fontsize': 16,
                    'font.size': 16,
                    'font.family': 'sans-serif',
                    'font.sans-serif': 'Arial',
                    'mathtext.fontset': 'dejavusans',
                    'axes.linewidth': 1,
                    'lines.linewidth': 1.,
                    'lines.markersize': 1.})
fig, ax = plt.subplots(figsize = (6, 5))
ax.scatter(demands[0:2], dRatios[0:2], c = ['black', 'slategray'], alpha = 0.75, marker='D', s = 50)
ax.scatter(demands[2:4], dRatios[2:4], c = ['black', 'slategray'], alpha = 0.75, marker='s', s = 50)
ax.scatter(demands[4:6], dRatios[4:6], c = ['black', 'slategray'], alpha = 0.75, marker='X', s = 50)
ax.scatter(demands[6:8], dRatios[6:8], c = ['black', 'slategray'], alpha = 0.75, marker='P', s = 50)
ax.scatter(demands[8], dRatios[8], c = 'black', alpha = 0.75, marker='o', s = 50)
ax.set_xlabel("Maxmium Demand Charge [$/kW]")
ax.set_ylabel("Ratio of On : Off Peak Charge [-]")
ax.set_xlim(0.0, 50)
ax.set_ylim(0.0, 5.)
ax.set_xticks(np.arange(0.0, 60, 10))
ax.set_yticks(np.arange(0.0, 6., 1))
fig.tight_layout()
fig.savefig('demandCharge.pdf', dpi=300, bbox_inches='tight')
fig.savefig('demandCharge.svg', dpi=300, bbox_inches='tight')
fig.savefig('demandCharge.png', dpi=300, bbox_inches='tight')
```
